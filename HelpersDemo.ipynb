## Example 4
Load the dataset with a custom preprocessing function of choice. 
This example demonstrates interoperability with keras preprocessing functions.

````python
from keras.applications.imagenet_utils import preprocess_input
X, y, class_labels = helpers.load_dataset('/Users/swg/Desktop/dataset-aligned.nosync',
                                         preprocess=preprocess_input)
````

## Example 5
Another example of loading dataset with preprocessing. This example demonstrates user defined preprocessing functions.

````python
import cv2
def pp_image(img):
    return cv2.equalizeHist(img)
X, y, class_labels = helpers.load_dataset('/Users/swg/Desktop/dataset-aligned.nosync', 
                     preprocess=pp_image, gray=True)
````

## Example 6
This example demonstrates that images imread in grayscale can also be displayed using same helper function without any change in parameters.

````python
img = X[222]
helpers.cv2_imshow('Sample Image', img)
plt.show()
````
