```python
import cv2
```

```python
from matplotlib import pyplot as plt
%matplotlib inline
def show_image_inline(image):
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    plt.imshow(image)
```

```python
TEST_IMAGE = './data/testi1.jpg'
image = cv2.imread(TEST_IMAGE)
show_image_inline(image)
```

```python
hcc = cv2.CascadeClassifier('cascades/data/haarcascade_frontalface_alt2.xml')
grayed = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
faces = hcc.detectMultiScale(grayed, scaleFactor = 1.2, minNeighbors = 5)
print('Total', len(faces), 'faces were detected')
```

```python
rect_color = (0, 0, 255) # remember BGR, not RGB (0-255, regular)
rect_stroke = 2
for (x, y, w, h) in faces:
    cv2.rectangle(image, (x, y), (x + w, y + h), rect_color, rect_stroke)
```

```python
OUT_IMAGE = './data/testiout1.png'
show_image_inline(image)
cv2.imwrite(OUT_IMAGE, image)
```
